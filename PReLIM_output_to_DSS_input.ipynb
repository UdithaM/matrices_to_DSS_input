```python
def matrices_to_dss_input(matrix_dictionary):
    # pos = coordinate
    # N = total number of reads
    # X = number of reads showing methylation
    headers = ['chr','pos', 'N', 'X']
    chromosomes = []
    coordinates = []
    counts_methylated = []
    reads = []

    for key, value in matrix_dictionary.items():
        cpg_coords = list(value.columns)
        transpose = value.to_numpy().transpose() 
        for i in range(len(cpg_coords)):
            
            chromosomes.append(key.split('_')[0])
            coordinates.append(cpg_coords[i] + 1)
            num_unmethylated = (transpose[i] == 0).sum()
            num_methylated = (transpose[i] == 1).sum()
            reads.append(num_methylated + num_unmethylated)
            counts_methylated.append(num_methylated)
        
    
    list_of_tuples = list(zip(chromosomes, coordinates, reads, counts_methylated))
    df = pd.DataFrame(list_of_tuples, columns=headers)
    
    return df
```

```python
wbc3842_chr5_pre_imputation_df = matrices_to_dss_input(wbc3842_chr5_pre_impute_matrices)
wbc3842_chr5_pre_imputation_df[0:50]
```

```python
wbc3842_chr5_imputed_cpg2_df = matrices_to_dss_input(wbc3842_chr5_imputed_cpg2)
wbc3842_chr5_imputed_cpg3_df = matrices_to_dss_input(wbc3842_chr5_imputed_cpg3)
wbc3842_chr5_imputed_cpg4_df = matrices_to_dss_input(wbc3842_chr5_imputed_cpg4)
wbc3842_chr5_imputed_cpg5_df = matrices_to_dss_input(wbc3842_chr5_imputed_cpg5)

```

```python
# making cpg_index as the index of the df
wbc3842_chr5_pre_imputation_df = wbc3842_chr5_pre_imputation_df.reset_index(drop=True).set_index('pos', drop=False)

# making cpg_index as the index of the df as imputed ones are not sorted
wbc3842_chr5_imputed_cpg2_df = wbc3842_chr5_imputed_cpg2_df.reset_index(drop=True).set_index('pos', drop=False)
wbc3842_chr5_imputed_cpg2_df.sort_index(inplace=True)

# making cpg_index as the index of the df as imputed ones are not sorted
wbc3842_chr5_imputed_cpg3_df = wbc3842_chr5_imputed_cpg3_df.reset_index(drop=True).set_index('pos', drop=False)
wbc3842_chr5_imputed_cpg3_df.sort_index(inplace=True)

# making cpg_index as the index of the df as imputed ones are not sorted
wbc3842_chr5_imputed_cpg4_df = wbc3842_chr5_imputed_cpg4_df.reset_index(drop=True).set_index('pos', drop=False)
wbc3842_chr5_imputed_cpg4_df.sort_index(inplace=True)

# making cpg_index as the index of the df as imputed ones are not sorted
wbc3842_chr5_imputed_cpg5_df = wbc3842_chr5_imputed_cpg5_df.reset_index(drop=True).set_index('pos', drop=False)
wbc3842_chr5_imputed_cpg5_df.sort_index(inplace=True)
```

```python
# replacing pre-impute rows with imputed ones if available

wbc3842_chr5_pre_imputation_df.loc[wbc3842_chr5_pre_imputation_df['pos'].isin(wbc3842_chr5_imputed_cpg2_df['pos']), ['N', 'X']] = wbc3842_chr5_imputed_cpg2_df[['N', 'X']]
wbc3842_chr5_pre_imputation_df.loc[wbc3842_chr5_pre_imputation_df['pos'].isin(wbc3842_chr5_imputed_cpg3_df['pos']), ['N', 'X']] = wbc3842_chr5_imputed_cpg3_df[['N', 'X']]
wbc3842_chr5_pre_imputation_df.loc[wbc3842_chr5_pre_imputation_df['pos'].isin(wbc3842_chr5_imputed_cpg4_df['pos']), ['N', 'X']] = wbc3842_chr5_imputed_cpg4_df[['N', 'X']]
wbc3842_chr5_pre_imputation_df.loc[wbc3842_chr5_pre_imputation_df['pos'].isin(wbc3842_chr5_imputed_cpg5_df['pos']), ['N', 'X']] = wbc3842_chr5_imputed_cpg5_df[['N', 'X']]


wbc3842_chr5_replaced = wbc3842_chr5_pre_imputation_df
# df with cpgs having imputed reads replaced 
print(wbc3842_chr5_replaced.isna().sum())
wbc3842_chr5_replaced.head(30)
```
